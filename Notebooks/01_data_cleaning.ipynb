## Price Validation

### Objective

Investigate transactions with a unit price less than or equal to zero and determine whether they should be included in the sales analysis.

### Investigation

The dataset was filtered to identify all transactions where the unit price was less than or equal to zero.

Key findings:
- A number of records had a unit price of **0**.
- These records included a mixture of:
  - Inventory adjustment records (e.g., `damaged`, `lost`, `wet`, `MIA`)
  - Operational records (e.g., `Manual`, `POSTAGE`, `Discount`)
  - Placeholder descriptions (`?`)
  - Legitimate products with a unit price of zero.

### Observation

A unit price of zero does not always indicate an invalid record. Some zero-priced transactions represent operational activities or complimentary items rather than completed customer purchases.

Since the objective of this project is **sales analytics**, only revenue-generating transactions should be included in the sales dataset.

### Decision

The master cleaned dataset retains these records to preserve the original business information.

A separate `sales_df` was created by:
- Excluding cancelled transactions.
- Excluding transactions with a unit price less than or equal to zero.
- Excluding operational inventory records that do not represent customer sales.

This ensures that all analyses are performed only on completed, revenue-generating sales transactions.

```python
sales_df["TotalSales"] = sales_df["Quantity"] * sales_df["Price"]
```

```python
sales_df["Year"] = sales_df["InvoiceDate"].dt.year
sales_df["Month"] = sales_df["InvoiceDate"].dt.month
sales_df["MonthName"] = sales_df["InvoiceDate"].dt.month_name()
sales_df["Quarter"] = sales_df["InvoiceDate"].dt.quarter
sales_df["Weekday"] = sales_df["InvoiceDate"].dt.day_name()
sales_df["Hour"] = sales_df["InvoiceDate"].dt.hour
sales_df["IsWeekend"] = sales_df["Weekday"].isin(
    ["Saturday", "Sunday"]
)
```

## Feature Engineering

### Objective

Create additional features from the existing transactional data to simplify analysis and enable more meaningful business insights.

### Features Created

| Feature | Description |
|---------|-------------|
| TotalSales | Total revenue generated for each transaction (`Quantity × Price`) |
| Year | Year of the transaction |
| Month | Month number of the transaction |
| MonthName | Name of the transaction month |
| Quarter | Quarter in which the transaction occurred |
| Weekday | Day of the week of the transaction |
| Hour | Hour at which the transaction occurred |
| IsWeekend | Indicates whether the transaction occurred on a weekend |
